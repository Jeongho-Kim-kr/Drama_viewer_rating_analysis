```python
import datetime # 날짜

days_range=[]

start=datetime.datetime.strptime("2020-01-01", "%Y-%m-%d")
end=datetime.datetime.strptime("2021-01-01", "%Y-%m-%d")
date_generated=[start+datetime.timedelta(days=x) for x in range(0, (end-start).days)]

for date in date_generated:
    days_range.append(date.strftime("%Y-%m-%d"))
```

```python
days=[]

for day in days_range:
    days.append(day.replace('-',''))
```

```python
import pandas as pd # 크롤링
import urllib.request
import urllib.parse
from bs4 import BeautifulSoup
import re

date, date1, date2, date3=[], [], [], []
chanel, chanel1, chanel2, chanel3=[], [], [], []
drama, drama1, drama2, drama3=[], [], [], []
rating, rating1, rating2, rating3=[], [], [], []

for day in days: # 지상파
    url=f'https://www.nielsenkorea.co.kr/tv_terrestrial_day.asp?menu=Tit_1&sub_menu=1_1&area=00&begin_date={day}'

    html=urllib.request.urlopen(url).read()
    soup=BeautifulSoup(html, 'html.parser')

    date_a=[day]*20
    for date_b in date_a:
        date1.append(date_b)

    broadcastings=soup.find_all("td",class_="tb_txt_center")
    del broadcastings[40:]

    for broadcasting in broadcastings:
        if  len(broadcasting.get_text()) > 3:
            chanel1.append(broadcasting.get_text().strip('\t'))

    titles=soup.find_all("td",class_="tb_txt")
    del titles[20:]

    for title in titles:
        drama1.append(title.get_text().strip('\t'))

    percents=soup.find_all("td",class_="percent")
    del percents[10:]

    for percent in percents:
        rating1.append(percent.get_text().replace('\r\n','').strip())

    percents_g=soup.find_all("td",class_="percent_g")
    del percents_g[10:]

    for percent_g in percents_g:
        rating1.append(percent_g.get_text().replace('\r\n','').strip()) 
```

```python
for day in days: # 종편
    url=f'https://www.nielsenkorea.co.kr/tv_terrestrial_day.asp?menu=Tit_1&sub_menu=2_1&area=00&begin_date={day}'

    html=urllib.request.urlopen(url).read()
    soup=BeautifulSoup(html, 'html.parser')

    date_a=[day]*10
    for date_b in date_a:
        date2.append(date_b)

    broadcastings=soup.find_all("td",class_="tb_txt_center")
    del broadcastings[20:]

    for broadcasting in broadcastings:
        if  len(broadcasting.get_text()) > 3:
            chanel2.append(broadcasting.get_text().strip('\t'))

    titles=soup.find_all("td",class_="tb_txt")
    del titles[10:]

    for title in titles:
        drama2.append(title.get_text().strip('\t'))

    percents=soup.find_all("td",class_="percent")
    del percents[10:]

    for percent in percents:
        rating2.append(percent.get_text().replace('\r\n','').strip())
```

```python
for day in days: # 케이블
    url=f'https://www.nielsenkorea.co.kr/tv_terrestrial_day.asp?menu=Tit_1&sub_menu=3_1&area=00&begin_date={day}'

    html=urllib.request.urlopen(url).read()
    soup=BeautifulSoup(html, 'html.parser')

    date_a=[day]*10
    for date_b in date_a:
        date3.append(date_b)

    broadcastings=soup.find_all("td",class_="tb_txt_center")
    del broadcastings[20:]

    for broadcasting in broadcastings:
        if  len(broadcasting.get_text()) > 3:
            chanel3.append(broadcasting.get_text().strip('\t'))

    titles=soup.find_all("td",class_="tb_txt")
    del titles[10:]

    for title in titles:
        drama3.append(title.get_text().strip('\t'))

    percents=soup.find_all("td",class_="percent")
    del percents[10:]

    for percent in percents:
        rating3.append(percent.get_text().replace('\r\n','').strip())
```

```python
date=date1+date2+date3
chanel=chanel1+chanel2+chanel3
drama=drama1+drama2+drama3
rating=rating1+rating2+rating3
        

dic={'날짜':date, '채널':chanel, '드라마 제목':drama, '시청률':rating}

my_df=pd.DataFrame(data=dic, columns=["날짜", "채널", "드라마 제목", "시청률"])
my_df
```

```python
index=[] # 드라마만 추출
for i in range(len(my_df)):
    if ('드라마' not in my_df['드라마 제목'][i]) and ('연속극' not in my_df['드라마 제목'][i]) and ('미니시리즈' not in my_df['드라마 제목'][i]) and ('주말특별기획' not in my_df['드라마 제목'][i]):
        index.append(i)
index.sort()
index.reverse()
for i in index:
    my_df.drop(my_df.index[i], inplace=True)

my_df=my_df.sort_values('날짜') # 정렬
my_df=my_df.reset_index(drop=True) 
my_df
```

```python
my_df.to_excel("시청률.xlsx")
```
