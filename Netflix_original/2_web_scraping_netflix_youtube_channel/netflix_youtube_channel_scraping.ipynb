```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
from bs4 import BeautifulSoup
```

```python
def get_titles():
    titles = soup.find_all('a', attrs = {'class':'yt-simple-endpoint style-scope ytd-grid-video-renderer'})
    titles = [element.get_text() for element in titles]
    return titles
```

```python
def get_viewers():
    viewers_date = soup.find_all('span', attrs = {'class':'style-scope ytd-grid-video-renderer'})
    viewers_date = [element.get_text() for element in viewers_date]
    viewers = viewers_date[0:len(viewers_date):2]
    viewers = [element.replace('조회수 ','') for element in viewers]
    return viewers
```

```python
def get_date():
    viewers_date = soup.find_all('span', attrs = {'class':'style-scope ytd-grid-video-renderer'})
    viewers_date = [element.get_text() for element in viewers_date]
    date = viewers_date[1:len(viewers_date):2]
    date = [element.replace('스트리밍 시간: ','') for element in date]
    date = [element.replace('시간 전','H') for element in date]
    date = [element.replace('일 전','D') for element in date]
    date = [element.replace('주 전','W') for element in date]
    date = [element.replace('개월 전','M') for element in date]
    date = [element.replace('년 전','Y') for element in date]
    return date
```
