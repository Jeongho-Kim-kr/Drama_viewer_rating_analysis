```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import requests
from bs4 import BeautifulSoup
```

```python
def get_titles():
    titles = soup.find_all('a', attrs = {'class':'yt-simple-endpoint style-scope ytd-grid-video-renderer'})
    titles = [element.get_text() for element in titles]
    return titles
```

```python
def get_viewers():
    viewers_date = soup.find_all('span', attrs = {'class':'style-scope ytd-grid-video-renderer'})
    viewers_date = [element.get_text() for element in viewers_date]
    viewers = viewers_date[0:len(viewers_date):2]
    viewers = [element.replace('Ï°∞ÌöåÏàò ','') for element in viewers]
    return viewers
```

```python
def get_date():
    viewers_date = soup.find_all('span', attrs = {'class':'style-scope ytd-grid-video-renderer'})
    viewers_date = [element.get_text() for element in viewers_date]
    date = viewers_date[1:len(viewers_date):2]
    date = [element.replace('Ïä§Ìä∏Î¶¨Î∞ç ÏãúÍ∞Ñ: ','') for element in date]
    date = [element.replace('ÏãúÍ∞Ñ Ï†Ñ','H') for element in date]
    date = [element.replace('Ïùº Ï†Ñ','D') for element in date]
    date = [element.replace('Ï£º Ï†Ñ','W') for element in date]
    date = [element.replace('Í∞úÏõî Ï†Ñ','M') for element in date]
    date = [element.replace('ÎÖÑ Ï†Ñ','Y') for element in date]
    return date
```
